```python
import pandas as pd
import matplotlib.pyplot as plt

#
url = 'https://raw.githubusercontent.com/rubensuarez22/DataSetML/refs/heads/main/Air_Traffic_Passenger_Statistics.csv'

df = pd.read_csv(url)

df.head()

df.head(25)
```

```python

# Step 1: Eliminate all the columns leaving only: "Activity Period", "Operating Airline","Published Airline IATA Code", "GEO Region", "Activity Type Code", and "Passenger Count"

columns_to_keep = ["Activity Period", "Operating Airline", "Published Airline IATA Code", "GEO Region", "Activity Type Code", "Passenger Count"]
df_filtered = df[columns_to_keep]
df_filtered.head()

print('Lets see how many many rows we have filter the rows that doesnt have a AITA Code:', df_filtered.shape[0])
```

```python
# Step 2: Eliminate all the rows that do not have a "Published Airline IATA Code"
df_filtered = df_filtered.dropna(subset=["Published Airline IATA Code"])

print('Now we reduced the number to', df_filtered.shape[0])
```

```python
# Step 3 Substitute the "GEO Region" values with a numerical code, explaining why you decided to choose a certain code.

# Ver la frecuencia de los valores en la columna "GEO Region"
print(df_filtered["GEO Region"].value_counts())
# Diccionaro para mapear las regiones.
geo_region_mapping = {
    "US": 1,
    "Asia":2,
    "Europe": 3,
    "Canada": 4,
    "Mexico": 5,
    "Australia / Oceania": 6,
    "Central America": 7,
    "Middle East": 8,
    "South America": 9,
}

#Los codigos fueron escogidos por la frecuencia
# Ver los valores únicos en la columna "GEO Region"

# Aplicar el mapeo a la columna "GEO Region"
df_filtered["GEO Region"] = df_filtered["GEO Region"].map(geo_region_mapping)

# Verificar los resultados
df_filtered.head()
```

```python
#Step 4: Create a new table and a graph for all flights operated by a mexican airline, showing one line for each category of the "Activity Type Code", showing how many flights have deplaned in San Francisco, how many enplaned, or passed though in transit in each month.


# Filtrar vuelos operados por aerolíneas mexicanas (ejemplo: Aeroméxico, Volaris, Interjet)
mexican_airlines_codes = ['AM', 'Y4', '4O']  # Lista de códigos IATA de aerolíneas mexicanas obtenidas de https://madavan.com.mx/codigo-iata-aerolineas/

df_mexican_airlines = df_filtered[df_filtered['Published Airline IATA Code'].isin(mexican_airlines_codes)]

# Mostrar el resultado filtrado
print(df_mexican_airlines.head())
print('Total de columnas con aerolineas mexicanas', df_mexican_airlines.shape[0])

```

```python
#Extra: For 2 extra points; calculate the average number of passengers per year that have Deplaned in this airport.


# Obtener los aterrizajes en 2015
df_deplaned_2015 = df_filtered[
    (df_filtered["Activity Type Code"] == "Deplaned") &
    (df_filtered["Activity Period"].astype(str).str[:4] == '2015')
]

# Sumar el conteo de pasajeros deplaned en 2015
total_deplaned_2015 = df_deplaned_2015['Passenger Count'].sum()
print("Total de pasajeros deplaned en 2015:", total_deplaned_2015)

```
